```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
games = ['LOL', 'Dota 2', 'Counter-Strike',
        'DayZ', 'HOTS', 'Isaac',
         'GTS', 'Hearthstone', 'World of Tanks',
         'Agar.io']
viewers = [1070, 472, 302, 239, 210, 171, 170, 90, 86, 71]
```

```python
fig = plt.figure(figsize=(12, 9))

sns.set_style('darkgrid')
sns.barplot(games, viewers)
plt.xlabel('Games', fontsize=20)
plt.ylabel('# of Viewers',fontsize=20)
plt.xticks(fontsize=14, rotation=-15)
plt.yticks(fontsize=14)
plt.savefig('Games_barchart.pdf');
```

```python
genre = ['MOBA', 'Other', 'FPS', 'Survival']
amts = [1752, 927, 302, 19]
explode = (.1, 0, 0, 0)
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.pie(amts, labels=genre, autopct='%.1f%%', explode=explode, shadow=True,
      textprops = {'fontsize':15})
plt.title('Twitch Genre Distribution', fontsize=20)
plt.savefig('genre_pie.pdf')
;
```
